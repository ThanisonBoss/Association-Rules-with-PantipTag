```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dataPath = "data/pantipTagFinal.csv"
data = pd.read_csv(dataPath)
data.head(10)
```

```python
data.shape
```

```python
dataList = list(data["itemsets"].apply(lambda x : x.strip("[]").replace("'","").split(",")))
dataList
```

```python
encoder = TransactionEncoder()
data = encoder.fit(dataList).transform(dataList)
data = pd.DataFrame(data, columns=encoder.columns_)
data = data.astype("int")
data
```

```python
dataSum = pd.DataFrame(data.sum().reset_index())
dataSum.columns = ["item","amount"]
dataSum = dataSum.sort_values("amount", ascending=False)
dataSum
```

```python
dataApriori = apriori(data, min_support=0.01, use_colnames=True, verbose=1)
dataApriori
```

```python
dataRules = association_rules(dataApriori, metric="confidence", min_threshold=0.5)
dataRules
```
